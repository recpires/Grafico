```python
import matplotlib.pyplot as plt
import numpy as np
```

### Gerando 1000 pares de números aleatórios entre 0 e 10

```python
x = np.random.uniform(0, 10, 1000)
y = np.random.uniform(0, 10, 1000)

# Criando um gráfico de dispersão
plt.scatter(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Dispersão de Números Aleatórios')
plt.show()
```
